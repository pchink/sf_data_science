```python
import pandas as pd
```

```python
movies_data = pd.read_csv('data/movies_data/movies.csv', sep=',')
movies_data.nunique
```

```python
ratings1 = pd.read_csv('data/movies_data/ratings1.csv', sep=',')
ratings1['userId'].nunique()
```

```python
# 37.5.4
dates = pd.read_csv('data/movies_data/dates.csv', sep=',')
dates['date'] = pd.to_datetime(dates['date'])
year = dates['date'].dt.year
year.mode()
```

```python
ratings2 = pd.read_csv('data/movies_data/ratings2.csv', sep=',')
ratings = pd.concat([ratings1, ratings2])
display(ratings)
```

```python
ratings = pd.concat(
    [ratings1, ratings2],
    ignore_index=True
)
display(ratings)
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])
```

```python
display(ratings1.tail(1))
display(ratings2.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```

```python
#объединим таблицы типом left
joined_false = ratings_dates.join(
    movies_data,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
#совмещаем таблицы по ключевому столбцу с помощью метода join(),
#используем ключевой столбец в «правой» таблице в качестве индекса. 
joined = ratings_dates.join(
    movies_data.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

```python
#объединим таблицы ratings1 и ratings2, как мы уже делали раньше, но теперь используем метод merge():
merge_ratings = ratings1.merge(ratings2, how='outer')
print('Число строк в таблице merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
# Число строк в таблице merge_ratings: 100836
```
