```python
#  Задание 36.6.10
#  В качестве ответа запишите среднюю длительность поездки в секундах.

bike['stoptime'] = pd.to_datetime(bike['stoptime'])
bike['starttime'] = pd.to_datetime(bike['starttime'])
bike['trip duration'] = (bike['stoptime'] - bike['starttime']).dt.seconds
round(bike['trip duration'].mean(), 2)
```

```python
# Задание 36.6.11
# Выясните, сколько поездок начиналось в выходные.
bike['starttime'] = pd.to_datetime(bike['starttime'])
bike['weekend'] = bike['starttime'].dt.dayofweek
weekend_count = bike[(bike['weekend'] == 5) | (bike['weekend'] == 6)].shape[0]
weekend_count
```

```python
# 36.6.12
# Во сколько раз количество поездок, совершённых днём, больше, чем количество поездок, 
# совёршенных ночью, за представленный в данных период времени?
def get_time_of_day(time):
    if 0 <= time <= 6:
        return 'night'
    elif 6 < time <= 12:
        return 'morning'
    elif 12 < time <= 18:
        return 'day'
    elif 18 < time <= 23:
        return 'evening'
    else:
        return 'else'
bike['time_of_day'] = bike['starttime'].dt.hour.apply(get_time_of_day)
a = bike[bike['time_of_day'] == 'day'].shape[0]
b = bike[bike['time_of_day'] == 'night'].shape[0]
print(round(a / b))
```
